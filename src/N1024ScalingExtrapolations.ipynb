### 3-parameter exponential fitted $\beta$-scaling

Want to fit observable data as a function of $\beta$ to the form:
$$
\mathcal{O}(\beta) = \mathcal{O}_{\beta\to\infty} + C_1 e^{-C_2 \beta}
$$

where $\mathcal{O}_{\beta\to\infty}$,$C_1$, and $C_2$ are fitting parameters. The $\mathcal{O}_{\beta\to\infty}$ parameter will give a large $\beta$ extrapolation of the observable in question. Here, we will focus on the second RÃ©nyi entanglement entropy $S_2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from scipy import optimize
```

Define the exponential model

```python
def exponential(x, c1, c2, c3):
    return c3 + c1*np.exp(-x*c2)
```

```python
# Array where extrapolated S2s will be saved to
extrapolated_S2s = np.zeros(20)
extrapolated_S2s_err = np.zeros(20)
```

Feed data points (obtained from a separate script)

```python
betas = np.array([2,4,6,8,10,12,16,20,24]) # betas
U = 16.666667
```

```python
data1 = np.loadtxt("../ProcessedData/2D_32_1024_1_16.666667_1.000000_betas_10001_S2.dat")
data2 = np.loadtxt("../ProcessedData/2D_32_1024_2_16.666667_1.000000_betas_10001_S2.dat")
data3 = np.loadtxt("../ProcessedData/2D_32_1024_3_16.666667_1.000000_betas_10001_S2.dat")
data4 = np.loadtxt("../ProcessedData/2D_32_1024_4_16.666667_1.000000_betas_10001_S2.dat")
data5 = np.loadtxt("../ProcessedData/2D_32_1024_5_16.666667_1.000000_betas_10001_S2.dat")
data6 = np.loadtxt("../ProcessedData/2D_32_1024_6_16.666667_1.000000_betas_10001_S2.dat")
data7 = np.loadtxt("../ProcessedData/2D_32_1024_7_16.666667_1.000000_betas_10001_S2.dat")
data8 = np.loadtxt("../ProcessedData/2D_32_1024_8_16.666667_1.000000_betas_10001_S2.dat")
data9 = np.loadtxt("../ProcessedData/2D_32_1024_9_16.666667_1.000000_betas_10001_S2.dat")
data10 = np.loadtxt("../ProcessedData/2D_32_1024_10_16.666667_1.000000_betas_10001_S2.dat")
data11 = np.loadtxt("../ProcessedData/2D_32_1024_11_16.666667_1.000000_betas_10001_S2.dat")
data12 = np.loadtxt("../ProcessedData/2D_32_1024_12_16.666667_1.000000_betas_10001_S2.dat")
data13 = np.loadtxt("../ProcessedData/2D_32_1024_13_16.666667_1.000000_betas_10001_S2.dat")
data14 = np.loadtxt("../ProcessedData/2D_32_1024_14_16.666667_1.000000_betas_10001_S2.dat")
data15 = np.loadtxt("../ProcessedData/2D_32_1024_15_16.666667_1.000000_betas_10001_S2.dat")
data16 = np.loadtxt("../ProcessedData/2D_32_1024_16_16.666667_1.000000_betas_10001_S2.dat")
data17 = np.loadtxt("../ProcessedData/2D_32_1024_17_16.666667_1.000000_betas_10001_S2.dat")
data18 = np.loadtxt("../ProcessedData/2D_32_1024_18_16.666667_1.000000_betas_10001_S2.dat")
data19 = np.loadtxt("../ProcessedData/2D_32_1024_19_16.666667_1.000000_betas_10001_S2.dat")
data20 = np.loadtxt("../ProcessedData/2D_32_1024_20_16.666667_1.000000_betas_10001_S2.dat")
```

```python
# Obtained with entanglement_combiner.py for betas
â = 1
y_array_exp = data1[:,1]
sigmas = data1[:,2]

# Perform 3-parameter exponential fit
popt_exponential, pcov_exponential = scipy.optimize.curve_fit(exponential, betas[0:], y_array_exp[0:],sigma=sigmas[0:],p0=[0.9,0.1,y_array_exp[-1]])

# Extract fit parameters
c1,c2,c3 = popt_exponential
c3_err = np.sqrt(np.diag(pcov_exponential)[2])

extrapolated_S2s[0] = c3
extrapolated_S2s_err[0] = c3_err

# Perform the fit
betas_fit = np.linspace(min(betas),max(betas)*1,1000)
y_fit = exponential(betas_fit,c1,c2,c3)

plt.plot(betas_fit,y_fit,label='fit')
plt.errorbar(betas,y_array_exp, yerr=sigmas,color='red',marker='o',label='data',fmt='.',
            capsize=5)
plt.ylabel(r"$S_2$")
plt.xlabel(r"$\beta$")
plt.legend(frameon=False);

print("S2 (extrapolated) \n")
print(f"{c3:0.6f} +/- {c3_err:0.6f}")
```
